En este caso se selecciona el método de rangos intercuartiles y límites de movilidad (IQR con factor 1.5) para la gestión de atípicos.

```python
# Se calcula el primer cuartil (Q1) de la columna 'rentabilidad_mensual'.
q1 = df['rentabilidad_mensual'].quantile(0.25)

# Se calcula el tercer cuartil (Q3) de la columna 'rentabilidad_mensual'.
q3 = df['rentabilidad_mensual'].quantile(0.75)

# Se calcula el rango intercuartílico (IQR) como la diferencia entre Q3 y Q1.
iqr = q3 - q1

# Se calcula el límite superior como Q3 más 1.5 veces el rango intercuartílico.
limite_superior = q3 + 1.5 * iqr

# Se calcula el límite inferior como Q1 menos 1.5 veces el rango intercuartílico.
limite_inferior = q1 - 1.5 * iqr

# Se crea un filtro booleano para identificar valores atípicos (outliers) en la columna 'rentabilidad_mensual'.
# Se marcan como outliers aquellos valores menores al límite inferior o mayores al límite superior.
filtro_outliers = (df['rentabilidad_mensual'] < limite_inferior) | (df['rentabilidad_mensual'] > limite_superior)

# Se imprime el valor calculado del límite superior de la detección de outliers.
print(f'El límite superior es: {limite_superior}')

# Se imprime el valor calculado del límite inferior de la detección de outliers.
print(f'El límite inferior es: {limite_inferior}')

# Se imprime la cantidad de valores que cumplen con el criterio de ser considerados outliers.
print(f'La cantidad de valores atípicos son: {filtro_outliers.sum()}')
```

```python
# Se eliminan los registros considerados como outliers en el DataFrame 'df' utilizando el filtro 'filtro_outliers'
df_sin_outliers = df[~filtro_outliers]

# Se obtienen las dimensiones (número de filas y columnas) del DataFrame resultante
df_sin_outliers.shape
```
